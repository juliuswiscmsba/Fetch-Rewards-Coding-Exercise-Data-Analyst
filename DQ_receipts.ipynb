```python
#Load library
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Read file
receipts = pd.read_json('receipts.json', lines=True)
```

```python
receipts.head()
```

```python
receipts.info()
```

```python
#Check missing value
receipts.isna().sum()/len(receipts)
```

```python
#Deal woth data format/type issue
#id(dict -> string)
new_id = []
id_list = receipts['_id'].tolist()
for i in id_list:
    new_id.append(i['$oid'])
    
receipts['_id'] = new_id 
receipts['_id'] = receipts['_id'].astype('string')
```

```python
#Date(object to datetime)
for col in ['createDate','dateScanned','finishedDate','modifyDate','pointsAwardedDate','purchaseDate']:
    datetemp=[]
    temp = receipts[col].tolist()
    for row in temp:
        if type(row) != float:
            datetemp.append(pd.to_datetime(row['$date']))
        else:
            datetemp.append('')
    receipts[col]=datetemp
```

```python
receipts.info()
```

```python
#rewardsReceiptItemList(dict -> new dataframe)
#Create a single table for receipt items becasue it will be too messy for the receipt table if we read the value(dict) in the 'rewardsReceiptItemList'.
receiptItem = pd.DataFrame()
for row in range(len(receipts['rewardsReceiptItemList'])):
    items = receipts['rewardsReceiptItemList'][row]
    receipt_id = pd.DataFrame({'receipt_id':receipts['_id'][row]},index=[0])
    if type(items) != float:
        for i in items:
            temp = receipt_id.join(pd.DataFrame.from_dict(i, orient='index').transpose())
            receiptItem = pd.concat([receiptItem, temp])
            
receiptItem.reset_index(drop = True, inplace=True)
receiptItem.head()
```

```python
#Drop rewardsReceiptItemList in receipts table to see whether there is any duplicate value.
receipts_without_item = receipts.drop('rewardsReceiptItemList', axis=1)
receipts_without_item.duplicated().sum()
```
